```python
import yolox
print("yolox.__version__:", yolox.__version__)
# from yolox.tracker.byte_tracker import BYTETracker, STrack
from onemetric.cv.utils.iou import box_iou_batch
from dataclasses import dataclass
@dataclass(frozen=True)
class BYTETrackerArgs:
    track_thresh: float = 0.25
    track_buffer: int = 30
    match_thresh: float = 0.8
    aspect_ratio_thresh: float = 3.0
    min_box_area: float = 1.0
    mot20: bool = False
from IPython import display
display.clear_output()
import supervision
print("supervision.__version__:", supervision.__version__)
from supervision.draw.color import ColorPalette
# settings
MODEL = "yolov8x.pt"
from ultralytics import YOLO
model = YOLO(MODEL)
model.fuse()
CLASS_NAMES_DICT = model.model.names
```

```python
# class_ids of interest - car, motorcycle, bus and truck
CLASS_ID = [2, 3, 5, 7]
from supervision.tools.detections import Detections, BoxAnnotator
import cv2

box_annotator = BoxAnnotator(color=ColorPalette(), thickness=1, text_thickness=1, text_scale=0.5)
dict_list = []

def process_images():
    image_paths = [r"1.jpg", r"2.jpg", r"3.jpg", r"img5.jpg"]
    images = [cv2.imread(path) for path in image_paths]

    for image in images:
        # Model prediction on single image and conversion to supervision Detections
        results = model(image)
        detections = Detections(
            xyxy=results[0].boxes.xyxy.cpu().numpy(),
            confidence=results[0].boxes.conf.cpu().numpy(),
            class_id=results[0].boxes.cls.cpu().numpy().astype(int)
        )

        object_counts = {class_name: 0 for class_name in CLASS_NAMES_DICT.values()}
        labels = [
            f"{CLASS_NAMES_DICT[class_id]} {confidence:0.2f}"
            for _, confidence, class_id, tracker_id
            in detections
        ]

        temp_dict = {}  # Create a new dictionary for each image
        for _, confidence, class_id, tracker_id in detections:
            class_name = CLASS_NAMES_DICT[class_id]
            if class_id in [2, 3, 5, 7]:
                object_counts[class_name] = object_counts.get(class_name, 0) + 1

        for class_id in [2, 3, 5, 7]:
            class_name = CLASS_NAMES_DICT[class_id]
            count = object_counts.get(class_name, 0)
            print(f"Number of {class_name}: {count}")
            temp_dict[class_name] = count
            print(temp_dict)

        dict_list.append(temp_dict)  # Append the new dictionary to dict_list

        # Annotate and display the image
        annotated_image = box_annotator.annotate(frame=image, detections=detections, labels=labels)
        cv2.imshow("Image",annotated_image)
        cv2.waitKey()

    return dict_list

print(process_images())
```

```python
import tkinter as tk


class TrafficLightGUI:

    def __init__(self, master):
        self.master = master
        self.master.title("Traffic Lights")

        self.canvas = tk.Canvas(self.master, width=998, height=500, bg="white")
        self.canvas.pack()

        self.red_light_1 = self.canvas.create_oval(50, 50, 150, 150, fill="gray")
        self.yellow_light_1 = self.canvas.create_oval(
            50, 50 + 130, 150, 150 + 130, fill="gray"
        )
        self.green_light_1 = self.canvas.create_oval(
            50, 50 + 2 * 130, 150, 150 + 2 * 130, fill="gray"
        )

        self.red_light_2 = self.canvas.create_oval(250, 50, 350, 150, fill="gray")
        self.yellow_light_2 = self.canvas.create_oval(
            250, 50 + 130, 350, 150 + 130, fill="gray"
        )
        self.green_light_2 = self.canvas.create_oval(
            250, 50 + 2 * 130, 350, 150 + 2 * 130, fill="gray"
        )

        self.red_light_3 = self.canvas.create_oval(450, 50, 550, 150, fill="gray")
        self.yellow_light_3 = self.canvas.create_oval(
            450, 50 + 130, 550, 150 + 130, fill="gray"
        )
        self.green_light_3 = self.canvas.create_oval(
            450, 50 + 2 * 130, 550, 150 + 2 * 130, fill="gray"
        )

        self.red_light_4 = self.canvas.create_oval(650, 50, 750, 150, fill="gray")
        self.yellow_light_4 = self.canvas.create_oval(
            650, 50 + 130, 750, 150 + 130, fill="gray"
        )
        self.green_light_4 = self.canvas.create_oval(
            650, 50 + 2 * 130, 750, 150 + 2 * 130, fill="gray"
        )

        self.red_label_1 = tk.Label(self.master, text="Red Light")
        self.red_label_1.place(x=100, y=30)
        self.yellow_label_1 = tk.Label(self.master, text="Yellow Light")
        self.yellow_label_1.place(x=100, y=160)
        self.green_label_1 = tk.Label(self.master, text="Green Light")
        self.green_label_1.place(x=100, y=290)

        self.red_label_2 = tk.Label(self.master, text="Red Light")
        self.red_label_2.place(x=300, y=30)
        self.yellow_label_2 = tk.Label(self.master, text="Yellow Light")
        self.yellow_label_2.place(x=300, y=160)
        self.green_label_2 = tk.Label(self.master, text="Green Light")
        self.green_label_2.place(x=300, y=290)

        self.red_label_3 = tk.Label(self.master, text="Red Light")
        self.red_label_3.place(x=500, y=30)
        self.yellow_label_3 = tk.Label(self.master, text="Yellow Light")
        self.yellow_label_3.place(x=500, y=160)
        self.green_label_3 = tk.Label(self.master, text="Green Light")
        self.green_label_3.place(x=500, y=290)

        self.red_label_4 = tk.Label(self.master, text="Red Light")
        self.red_label_4.place(x=700, y=30)
        self.yellow_label_4 = tk.Label(self.master, text="Yellow Light")
        self.yellow_label_4.place(x=700, y=160)
        self.green_label_4 = tk.Label(self.master, text="Green Light")
        self.green_label_4.place(x=700, y=290)

        # Instance variables to hold the state of each light
        self.red_1 = True
        self.yellow_1 = False
        self.green_1 = False

        self.red_2 = True
        self.yellow_2 = False
        self.green_2 = False

        self.red_3 = True
        self.yellow_3 = False
        self.green_3 = False

        self.red_4 = True
        self.yellow_4 = False
        self.green_4 = False

        self.street = -1

        self.change_lights()

    def change_lights(self):
        self.update_lights()

        self.master.after(500, self.change_red)

    def change_yellow(self):
        self.update_lights(yellow=True)
        # time.sleep(3)
        self.master.after(3000, self.change_red)

    def change_green(self, greenTime):
        self.update_lights(red=False, yellow=False, green=True)
        # time.sleep(greenTime/1000)
        self.master.after(greenTime, self.change_yellow)

    def change_red(self):
        self.update_lights(red=True)
        # time.sleep(3)
        self.street = (self.street + 1) % 4
        self.master.after(1, self.trafficAlg)

    def update_lights(self, red=False, yellow=False, green=False):
        if red:
            if self.street == 0:
                self.red_1 = True
                self.yellow_1 = False
                self.green_1 = False
            if self.street == 1:
                self.red_2 = True
                self.yellow_2 = False
                self.green_2 = False
            if self.street == 2:
                self.red_3 = True
                self.yellow_3 = False
                self.green_3 = False
            if self.street == 3:
                self.red_4 = True
                self.yellow_4 = False
                self.green_4 = False

        elif yellow:
            if self.street == 0:
                self.red_1 = False
                self.yellow_1 = True
                self.green_1 = False
            if self.street == 1:
                self.red_2 = False
                self.yellow_2 = True
                self.green_2 = False
            if self.street == 2:
                self.red_3 = False
                self.yellow_3 = True
                self.green_3 = False
            if self.street == 3:
                self.red_4 = False
                self.yellow_4 = True
                self.green_4 = False
        elif green:
            if self.street == 0:
                self.red_1 = False
                self.yellow_1 = False
                self.green_1 = True
            if self.street == 1:
                self.red_2 = False
                self.yellow_2 = False
                self.green_2 = True
            if self.street == 2:
                self.red_3 = False
                self.yellow_3 = False
                self.green_3 = True
            if self.street == 3:
                self.red_4 = False
                self.yellow_4 = False
                self.green_4 = True

        # Update the GUI based on the state of each light
        self.canvas.itemconfig(self.red_light_1, fill="red" if self.red_1 else "gray")
        self.canvas.itemconfig(
            self.yellow_light_1, fill="yellow" if self.yellow_1 else "gray"
        )
        self.canvas.itemconfig(
            self.green_light_1, fill="green" if self.green_1 else "gray"
        )

        self.canvas.itemconfig(self.red_light_2, fill="red" if self.red_2 else "gray")
        self.canvas.itemconfig(
            self.yellow_light_2, fill="yellow" if self.yellow_2 else "gray"
        )
        self.canvas.itemconfig(
            self.green_light_2, fill="green" if self.green_2 else "gray"
        )

        self.canvas.itemconfig(self.red_light_3, fill="red" if self.red_3 else "gray")
        self.canvas.itemconfig(
            self.yellow_light_3, fill="yellow" if self.yellow_3 else "gray"
        )
        self.canvas.itemconfig(
            self.green_light_3, fill="green" if self.green_3 else "gray"
        )

        self.canvas.itemconfig(self.red_light_4, fill="red" if self.red_4 else "gray")
        self.canvas.itemconfig(
            self.yellow_light_4, fill="yellow" if self.yellow_4 else "gray"
        )
        self.canvas.itemconfig(
            self.green_light_4, fill="green" if self.green_4 else "gray"
        )

    def trafficAlg(self): # need to add dict_list as a parameter
        greenTime = 0
        if dict_list[self.street]["car"]:
            greenTime += (
                dict_list[self.street]["car"] * 2000
            )  # we dunno car time and all but assuming
        if dict_list[self.street]["motorcycle"]:
            greenTime += (
                dict_list[self.street]["motorcycle"] * 1000
            )  # we dunno car time and all but assuming
        if dict_list[self.street]["bus"]:
            greenTime += (
                dict_list[self.street]["bus"] * 3000
            )  # we dunno car time and all but assuming
        if dict_list[self.street]["truck"]:
            greenTime += (
                dict_list[self.street]["truck"] * 4000
            )  # we dunno car time and all but assuming
        self.change_green(greenTime)


def main():
    root = tk.Tk()
    app = TrafficLightGUI(root)
    root.mainloop()


if __name__ == "__main__":
    main()
```
